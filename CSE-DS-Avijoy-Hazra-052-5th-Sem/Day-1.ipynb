```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('seaborn')
```

```python
x=np.linspace(0,10,20)
x
```

```python
y=np.square(x)
```

```python
y
```

```python
y1=x**3
```

```python
plt.figure(dpi=500)
plt.plot(x,y,color='red')
plt.plot(x,y1)
plt.plot(x,x**2+y**2)
plt.xlabel("X value")
plt.ylabel("Y value")
plt.title("Square")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,1)
plt.plot(y,x,'g')
plt.show()
```
