```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sal=pd.read_csv("Salaries.csv")
```

```python
sal.info()
```

```python
sal.head()
```

```python
def rem(x):
    if x==0 or x=='Not Provided' or x=='Not Provided' or x=='not provided':
        return None
    return x
```

```python
for col in sal.columns:
    sal[col]=sal[col].apply(rem)
```

```python
sal=sal.astype({'BasePay':np.float64,'OvertimePay':np.float64,'OtherPay':np.float64,'TotalPay':np.float64,'TotalPayBenefits':np.float64})
```

```python
sal.drop(columns=['Notes','Agency','Id'],inplace=True,axis=1)
```

```python
sal.info()
sal.to_csv('sal.csv')
```

```python
for i in sal.select_dtypes(np.float64).columns:
    sal[i]=sal[i].apply(np.abs)
```
